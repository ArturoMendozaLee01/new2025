```python
# prompt: widgets accordeon

import ipywidgets as widgets
from IPython.display import display, HTML

# Crea un acordeón con widgets
accordion = widgets.Accordion(children=[
    widgets.VBox([
        widgets.Label('Contenido del primer panel'),
        widgets.Text(placeholder='Ingrese texto')
    ]),
    widgets.VBox([
        widgets.Label('Contenido del segundo panel'),
        widgets.IntSlider()
    ])
])

# Define títulos para los paneles
accordion.set_title(0, 'Primer panel')
accordion.set_title(1, 'Segundo panel')

# Muestra el acordeón
display(accordion)
```
